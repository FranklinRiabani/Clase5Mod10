```python
from gensim.models import Word2Vec
import string
import os
from tqdm import tqdm
```

```python
# leer carpeta de nlp/Hielo y Fuego

path = 'nlp/Hielo y Fuego'

archivos=os.listdir(path)
archivos
```

```python
# unir todos los archivos en un solo texto

texto = ''
for archivo in archivos:
    with open(os.path.join(path, archivo), 'r', encoding='utf-8') as f:
        texto += f.read()
```

```python
len(texto)
```

```python
# dividir en oraciones

oraciones = texto.split('.') # dividir por punto o , o ; o : o ! o ?

len(oraciones)
```

```python
# imprimir las 10 primeras oraciones

for i in range(10):
    print(oraciones[i])
```

```python
# limpiar las oraciones
oraciones_limpias = []
# eliminar signos de puntuación
for oracion in oraciones:
    tokens = oracion.translate(str.maketrans('','',string.punctuation)).split()
    # print(tokens)
    #convertir a minúsculas
    tokens = [word.lower() for word in tokens if word.isalpha()]
    # print(tokens)
    if tokens:
        oraciones_limpias.append(tokens)
```

```python
oraciones_limpias
```

```python
# entrenar modelo word2vec

model=Word2Vec(sentences=oraciones_limpias,min_count=1,window=5,vector_size=500,workers=4)
```

```python
vector=model.wv['jon']
vector
```

```python
palabras_cercanas=model.wv.most_similar('dracarys',topn=5)
palabras_cercanas
```

```python
model.save('word2vec_hielo_y_fuego.model')
```

```python
# cargar modelo

model=Word2Vec.load('word2vec_hielo_y_fuego.model')
```

```python
#otros formatos de guardar el modelo

model.wv.save_word2vec_format('word2vec_hielo_y_fuego.txt',binary=False)
model.wv.save_word2vec_format('word2vec_hielo_y_fuego.bin',binary=True)
```

```python
from gensim.models import KeyedVectors

embeddings=KeyedVectors.load_word2vec_format('word2vec_hielo_y_fuego.txt',binary=False)

embeddings
```

```python
def analogia(a,b,c):
    resultado=embeddings.most_similar(positive=[a,c],negative=[b],topn=1)
    return resultado[0][0]
```

```python
print(analogia('jon','stark','lannister'))
```
